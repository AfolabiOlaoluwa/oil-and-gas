```python
scoutDF = scout.copy()
```

```python
scout = scoutDF[['NDIC File No:', 'API No:', 'Status Date:', 'Wellbore type:', 
                 'Latitude:', 'Longitude:', 'Total Depth:']]
```

```python
import numpy as np
def form_pt(data):
    #if not isinstance(data['Longitude:'], str):
    pt = 'Point(%s %s)' % (data['Longitude:'], data['Latitude:'])
    #else:
    #    pt = ''
    return pt
```

```python
scout['pt'] = scout.apply(form_pt, axis=1)
scout['pt'] = scout.pt.apply(lambda x: '' if x == 'Point( )' else x)
scout[1000:1010]
scout.head()
```

```python
scout[1000:1010]
```

```python
import uuid
uids = [[x, str(uuid.uuid4())] for x in list(scout['NDIC File No:'])]
uidDF = pd.DataFrame(uids, columns = ['NDIC File No:', 'uid'])
```

```python
scout = pd.merge(scout, uidDF, right_on = 'NDIC File No:', 
                     left_on = 'NDIC File No:')
scout.head()
```

```python
del scout['Latitude:']
del scout['Longitude:']
scout.columns = ['ndic', 'api', 'status_date', 'wellbore_type', 'total_depth', 'pt', 'uid']
```
