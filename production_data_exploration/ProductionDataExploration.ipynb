```python
import statsmodels.api as sm
fig, ax = plt.subplots(5, 5)
yearStart = 2014
numberPlots = ax.size
noLag['cumOil'] = noLag['cumOil'].astype(float)
#print numberPlots
ax = ax.ravel()
fits = []
for elem, year in enumerate(range(yearStart - 24, yearStart + 1)):
    temp = noLag[noLag.year == year]
    temp = temp.dropna()
    temp['ones']  = 1
    X = np.asarray(temp[['oil', 'ones']])
    #print X
    y = np.asarray(temp['cumOil'])
    #print y
    model = sm.OLS(y, X)
    results = model.fit()
    ax[elem].scatter(temp.oil, temp.cumOil)
    yPred = results.predict(X)
    ax[elem].plot(X[:,0], yPred)
    ax[elem].set_title(year)
    fit = list(results.params) + [year]
    fits.append(fit)
fig.set_size_inches(12, 12)
fit = pd.DataFrame(fits, columns = ['slope', 'offset', 'year'])
    
```

```python
## normalized histograms by total number of wells that ear.  Normalize by total number of wells, and
## share x and share y.  Use 5000 as the unit of measure.
fig, ax = plt.subplots(12, 1)
fig.set_size_inches(8, 20)
yearStart = 2014
ax = ax.ravel()
numPlots = ax.size
years = range(yearStart - numPlots + 1, yearStart + 1)
minVal = noLag[noLag.year.isin(years)].oil.min()
maxVal = noLag[noLag.year.isin(years)].oil.max()
binCalc = np.ceil((maxVal - minVal)/5000)
for elem, year in enumerate(years):
    temp = noLag[noLag.year == year]
    hist, bin_edges = np.histogram(temp.oil, bins = binCalc, range = (minVal, maxVal), density=True)
    hist = hist * 5000
    numWells = len(temp)
    ax[elem].text(maxVal/2, max(hist)-0.1, '%s, num_wells = %s' % (year, numWells))
    ax[elem].bar(bin_edges[:-1], hist, width = 1)
    ax[elem].set_xlim(min(bin_edges) -5000, max(bin_edges))
    
    
```
